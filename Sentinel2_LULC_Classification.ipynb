```python
import os

import cv2

import matplotlib.pyplot as plt

from ultralytics import YOLO

from google.colab import files, output
#Load the pre-trained Classification model

model = YOLO('yolov8n-cls.pt')

print("Upload the imgae .")
```

```python
# Cell 3


output.clear()#clear the previous output



print("Upload the Image")

uploaded = files.upload()



if uploaded:

    # Find the uploaded file name
    filename = list(uploaded.keys())[0]



    # Run the inference

    results = model(filename)



    for r in results:

        # FFid the top identity aur confidence level

        top_idx = r.probs.top1

        label = r.names[top_idx]

        conf = r.probs.top1conf.item() * 100



        # Image Plotting

        img = cv2.imread(filename)

        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)



        plt.figure(figsize=(6, 6))

        plt.imshow(img_rgb)

        plt.title(f"Detected: {label} ({conf:.1f}%)", color='blue', fontweight='bold')

        plt.axis('off')

        plt.show()



        # Analysis Details

        print(f"\n--- ANALYSIS FOR: {filename} ---")

        print(f"‚úÖ Identity: {label}")

        print(f"üìä Confidence Score: {conf:.1f}%")

        print(f"üåç Geo Context: This is a 10m resolution Sentinel-2 patch.")



        # Delete the Temp file

        os.remove(filename)

else:

    print("No file uploaded.")
```
